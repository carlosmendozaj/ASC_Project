Algoritmo inicial - tradicional:
Para el piso actual, encontrar el piso más cercano a mí y extablecerlo como mi príximo destino y repetir hasta que ya no haya mas pisos en la lista

```python
# Lista de órdenes de pisos
pisos = [1, 8, 12, 26, 5, 39]

# Piso actual
piso_actual = 10

# Inicializamos un contador para los pisos recorridos
pisos_recorridos_totales = 0

# Función que encuentra el piso más cercano al actual
def encontrar_piso_mas_cercano(piso_actual, pisos):
    # Encuentra el piso cuya diferencia con el piso actual sea mínima
    piso_mas_cercano = min(pisos, key=lambda x: abs(x - piso_actual))
    return piso_mas_cercano

# Algoritmo que sigue el proceso hasta que no haya más pisos en la lista
while pisos:
    # Encontrar el piso más cercano
    proximo_piso = encontrar_piso_mas_cercano(piso_actual, pisos)

    # Pisos recorridos:
    pisos_recorridos = abs(proximo_piso - piso_actual)
    pisos_recorridos_totales += pisos_recorridos

    # Resultados
    print(f"Voy del piso {piso_actual} al piso {proximo_piso} (recorro {pisos_recorridos} pisos)")

    # Establecer el próximo destino
    piso_actual = proximo_piso

    # Eliminar el piso visitado de la lista
    pisos.remove(proximo_piso)

# Total de pisos recorridos:
print(f"Pisos recorridos en total: {pisos_recorridos_totales}")
```

Algoritmo 2:
este algoritmo toma en cuenta a la dirección actual del ascensor (si va hacia arriba o hacia abajo) y sigue en esa dirección hasta que no queden más órdenes pendientes en esa dirección, antes de cambiar de rumbo. De esta manera, minimizamos los cambios de dirección, lo que debería reducir el costo total en pisos recorridos.

```python
# Lista de órdenes de pisos
ordenes = [1, 8, 12, 26, 5, 39]

# Piso actual
piso_actual = 10

# Inicializamos un contador para los pisos recorridos
pisos_recorridos_totales = 0

# Separar las órdenes en pisos por encima y por debajo del piso actual
pisos_arriba = sorted([piso for piso in ordenes if piso > piso_actual])
pisos_abajo = sorted([piso for piso in ordenes if piso < piso_actual], reverse=True)

# Imprimir las órdenes iniciales
print(f"Pisos arriba: {pisos_arriba}")
print(f"Pisos abajo: {pisos_abajo}")

# Primero vamos hacia arriba, si hay órdenes hacia arriba
for proximo_piso in pisos_arriba:
    pisos_recorridos = abs(proximo_piso - piso_actual)
    pisos_recorridos_totales += pisos_recorridos
    print(f"Voy del piso {piso_actual} al piso {proximo_piso} (recorro {pisos_recorridos} pisos)")
    piso_actual = proximo_piso

# Luego vamos hacia abajo, si hay órdenes hacia abajo
for proximo_piso in pisos_abajo:
    pisos_recorridos = abs(proximo_piso - piso_actual)
    pisos_recorridos_totales += pisos_recorridos
    print(f"Voy del piso {piso_actual} al piso {proximo_piso} (recorro {pisos_recorridos} pisos)")
    piso_actual = proximo_piso

# Imprimir el total de pisos recorridos
print(f"Pisos recorridos en total: {pisos_recorridos_totales}")
```

Algoritmo 3: TSP
El Problema del Viajante (TSP) busca encontrar la ruta más corta que visite un conjunto de ciudades (en este caso, pisos) y regrese al punto de origen.

```python
import itertools

# Lista de órdenes de pisos
ordenes = [1, 8, 12, 26, 5, 39]

# Piso actual
piso_actual = 10

# Función para calcular la distancia total recorrida para una ruta específica
def calcular_distancia_total(piso_actual, ruta):
    distancia_total = 0
    # Comienza desde el piso actual
    piso_anterior = piso_actual
    for piso in ruta:
        # Suma la distancia entre el piso anterior y el próximo piso en la ruta
        distancia_total += abs(piso - piso_anterior)
        piso_anterior = piso
    return distancia_total

# Algoritmo de Fuerza Bruta: Probar todas las permutaciones de los pisos
def tsp_fuerza_bruta(piso_actual, ordenes):
    # Encontrar todas las posibles permutaciones de los pisos a visitar
    permutaciones = itertools.permutations(ordenes)

    # Inicializar la mejor distancia como un valor muy grande
    mejor_distancia = float('inf')
    mejor_ruta = None

    # Probar todas las rutas posibles y encontrar la que tenga la menor distancia
    for ruta in permutaciones:
        distancia_actual = calcular_distancia_total(piso_actual, ruta)
        if distancia_actual < mejor_distancia:
            mejor_distancia = distancia_actual
            mejor_ruta = ruta

    return mejor_ruta, mejor_distancia

# Ejecutar el algoritmo TSP
mejor_ruta, mejor_distancia = tsp_fuerza_bruta(piso_actual, ordenes)

# Mostrar los resultados
print(f"La mejor ruta es: {mejor_ruta}")
print(f"La distancia total recorrida es: {mejor_distancia}")
```

```python
def algoritmo_scan(piso_actual, ordenes):
    # Dividimos las órdenes en dos listas: las que están por encima y por debajo del piso actual
    pisos_arriba = sorted([piso for piso in ordenes if piso > piso_actual])
    pisos_abajo = sorted([piso for piso in ordenes if piso < piso_actual], reverse=True)

    distancia_total = 0

    # Movemos hacia arriba primero
    for piso in pisos_arriba:
        distancia_total += abs(piso - piso_actual)
        piso_actual = piso

    # Movemos hacia abajo
    for piso in pisos_abajo:
        distancia_total += abs(piso - piso_actual)
        piso_actual = piso

    return distancia_total

# Ejemplo de uso
piso_actual = 10
ordenes = [1, 8, 12, 26, 5, 39]

distancia_total_scan = algoritmo_scan(piso_actual, ordenes)
print(f"Distancia total recorrida con SCAN: {distancia_total_scan}")
```

```python
def algoritmo_look(piso_actual, ordenes):
    # Dividimos las órdenes en dos listas: las que están por encima y por debajo del piso actual
    pisos_arriba = sorted([piso for piso in ordenes if piso > piso_actual])
    pisos_abajo = sorted([piso for piso in ordenes if piso < piso_actual], reverse=True)

    distancia_total = 0

    # Subimos hasta el último piso solicitado hacia arriba
    for piso in pisos_arriba:
        distancia_total += abs(piso - piso_actual)
        piso_actual = piso

    # Bajamos hasta el último piso solicitado hacia abajo
    for piso in pisos_abajo:
        distancia_total += abs(piso - piso_actual)
        piso_actual = piso

    return distancia_total

# Ejemplo de uso
distancia_total_look = algoritmo_look(piso_actual, ordenes)
print(f"Distancia total recorrida con LOOK: {distancia_total_look}")
```

Implementación de LOOK

```python
# Simulación de ascensores utilizando el algoritmo LOOK con control por piso
import random

# Tiempo promedio que tarda un ascensor por piso
tiempo_por_piso = 3  # en segundos
tiempo_puertas = 10  # tiempo que tarda en abrir/cerrar puertas

class Ascensor:
    def __init__(self, piso_actual):
        self.piso_actual = piso_actual
        self.direccion = "quieto"  # Puede ser 'subiendo', 'bajando' o 'quieto'

    def mover_a_piso(self, piso_destino):
        distancia = abs(self.piso_actual - piso_destino)
        self.piso_actual = piso_destino
        return distancia

def algoritmo_look_multiascensor(ascensores, solicitudes):
    """ Asigna ascensores y calcula tiempos usando el algoritmo LOOK para varios ascensores """
    total_tiempo_tradicional = 0
    total_tiempo_inteligente = 0

    # Procesamos las solicitudes
    for solicitud in solicitudes:
        piso_origen, piso_destino = solicitud

        # Sistema tradicional: seleccionamos el ascensor más cercano
        ascensor_tradicional = min(ascensores, key=lambda asc: abs(asc.piso_actual - piso_origen))
        distancia_tradicional = ascensor_tradicional.mover_a_piso(piso_origen)
        tiempo_tradicional = distancia_tradicional * tiempo_por_piso + tiempo_puertas
        total_tiempo_tradicional += tiempo_tradicional

        # Sistema inteligente (LOOK): seleccionamos el ascensor basado en el algoritmo LOOK
        ascensor_inteligente = min(ascensores, key=lambda asc: abs(asc.piso_actual - piso_origen))  # Similar criterio
        distancia_inteligente = ascensor_inteligente.mover_a_piso(piso_origen)
        tiempo_inteligente = distancia_inteligente * tiempo_por_piso + tiempo_puertas
        total_tiempo_inteligente += tiempo_inteligente

        # El ascensor inteligente luego se mueve al destino de forma óptima
        distancia_destino_inteligente = ascensor_inteligente.mover_a_piso(piso_destino)
        total_tiempo_inteligente += distancia_destino_inteligente * tiempo_por_piso + tiempo_puertas

        # El ascensor tradicional también se mueve al destino
        distancia_destino_tradicional = ascensor_tradicional.mover_a_piso(piso_destino)
        total_tiempo_tradicional += distancia_destino_tradicional * tiempo_por_piso + tiempo_puertas

    return total_tiempo_tradicional, total_tiempo_inteligente

# Crear ascensores en diferentes posiciones
ascensores = [Ascensor(piso_actual=random.randint(1, 20)) for _ in range(3)]

# Generar solicitudes (origen, destino) en los momentos clave del día
solicitudes = [(1, 15), (5, 18), (3, 10), (12, 20), (16, 5)]

# Ejecutar la simulación
tiempo_tradicional, tiempo_inteligente = algoritmo_look_multiascensor(ascensores, solicitudes)

print(f"Tiempo total en sistema tradicional: {tiempo_tradicional} segundos")
print(f"Tiempo total en sistema inteligente (LOOK): {tiempo_inteligente} segundos")
```

Caso 2 Simulador de capacidad del ascensor

```python
import numpy as np
import random

# Parámetros del ascensor
capacidad_max_personas = 10
capacidad_max_peso = 750  # en kg
tiempo_por_piso = 3  # en segundos
tiempo_puertas = 10  # en segundos

# Clase para personas que usan el ascensor
class Persona:
    def __init__(self, peso):
        self.peso = peso  # Peso de la persona

# Clase Ascensor que incluye la capacidad
class AscensorConCapacidad(Ascensor):
    def __init__(self, piso_actual):
        super().__init__(piso_actual)
        self.capacidad_personas = 0
        self.capacidad_peso = 0
        self.personas_a_bordo = []  # Lista de personas que están en el ascensor

    def cargar_personas(self, personas):
        """ Carga personas en el ascensor si hay suficiente capacidad """
        for persona in personas:
            if self.capacidad_personas + 1 <= capacidad_max_personas and self.capacidad_peso + persona.peso <= capacidad_max_peso:
                self.capacidad_personas += 1
                self.capacidad_peso += persona.peso
                self.personas_a_bordo.append(persona)
            else:
                print(f"Ascensor lleno. No se pudo subir una persona de {persona.peso} kg.")

    def descargar_personas(self):
        """ Descarga todas las personas del ascensor """
        num_personas = len(self.personas_a_bordo)
        total_peso = sum([persona.peso for persona in self.personas_a_bordo])
        self.capacidad_personas = 0
        self.capacidad_peso = 0
        self.personas_a_bordo = []
        print(f"Descargando {num_personas} personas con un total de {total_peso} kg.")

# Generar solicitudes con personas (peso aleatorio entre 50 kg y 100 kg)
def generar_solicitudes_con_personas(entrada_media, desviacion_estandar, num_personas):
    """
    Genera solicitudes con personas que tienen un peso específico.
    """
    solicitudes = []
    for _ in range(num_personas):
        piso_origen = int(np.random.normal(entrada_media, desviacion_estandar))
        piso_destino = int(np.random.normal(entrada_media + 5, desviacion_estandar))  # Añadimos desplazamiento para destino
        peso = random.randint(50, 100)  # Peso entre 50 kg y 100 kg
        persona = Persona(peso)
        if piso_origen != piso_destino:  # Evitar que el origen y destino sean el mismo
            solicitudes.append((piso_origen, piso_destino, persona))
    return solicitudes

# Algoritmo de simulación con capacidad
def algoritmo_ascensor_con_capacidad(ascensores, solicitudes):
    total_tiempo = 0

    # Procesamos las solicitudes
    for solicitud in solicitudes:
        piso_origen, piso_destino, persona = solicitud

        # Seleccionar el ascensor más cercano
        ascensor_asignado = min(ascensores, key=lambda asc: abs(asc.piso_actual - piso_origen))

        # Mover al piso de origen
        distancia_origen = ascensor_asignado.mover_a_piso(piso_origen)
        tiempo_origen = distancia_origen * tiempo_por_piso + tiempo_puertas
        total_tiempo += tiempo_origen

        # Intentar cargar a la persona
        ascensor_asignado.cargar_personas([persona])

        # Mover al destino
        distancia_destino = ascensor_asignado.mover_a_piso(piso_destino)
        tiempo_destino = distancia_destino * tiempo_por_piso + tiempo_puertas
        total_tiempo += tiempo_destino

        # Descargar personas en el destino
        ascensor_asignado.descargar_personas()

    return total_tiempo

# Simulación para un ascensor con capacidad
def simular_ascensor_con_capacidad():
    # Crear ascensores en diferentes posiciones
    ascensores = [AscensorConCapacidad(piso_actual=random.randint(1, 20)) for _ in range(3)]

    # Generar solicitudes con personas
    solicitudes = generar_solicitudes_con_personas(10, 3, 20)

    # Ejecutar la simulación
    tiempo_total = algoritmo_ascensor_con_capacidad(ascensores, solicitudes)

    print(f"Tiempo total con ascensores considerando capacidad: {tiempo_total} segundos")

# Ejecutar la simulación con ascensores que manejan capacidad
simular_ascensor_con_capacidad()
```

Agregando un ascensor prioritario:

```python
class AscensorPrioritario(Ascensor):
    """
    Clase para un ascensor prioritario. Solo atenderá solicitudes prioritarias.
    """
    def __init__(self, piso_actual):
        super().__init__(piso_actual)

# Generar solicitudes con usuarios prioritarios
def generar_solicitudes_prioritarias(entrada_media, desviacion_estandar, num_personas, prioridad_ratio=0.1):
    """
    Genera solicitudes de ingreso al ascensor, algunas con prioridad.
    """
    solicitudes = []
    for _ in range(num_personas):
        piso_origen = int(np.random.normal(entrada_media, desviacion_estandar))
        piso_destino = int(np.random.normal(entrada_media + 5, desviacion_estandar))  # Añadimos desplazamiento para destino
        es_prioritario = random.random() < prioridad_ratio  # Un porcentaje de solicitudes son prioritarias
        if piso_origen != piso_destino:  # Evitar que el origen y destino sean el mismo
            solicitudes.append((piso_origen, piso_destino, es_prioritario))
    return solicitudes

def algoritmo_look_multiascensor_prioritario(ascensores, ascensor_prioritario, solicitudes):
    """ Asigna ascensores y calcula tiempos usando el algoritmo LOOK para varios ascensores, incluyendo un ascensor prioritario """
    total_tiempo_tradicional = 0
    total_tiempo_inteligente = 0

    # Procesamos las solicitudes
    for solicitud in solicitudes:
        piso_origen, piso_destino, es_prioritario = solicitud

        # Si es prioritario, usar el ascensor prioritario
        if es_prioritario:
            ascensor_usado = ascensor_prioritario
        else:
            # Sistema tradicional: seleccionamos el ascensor más cercano
            ascensor_usado = min(ascensores, key=lambda asc: abs(asc.piso_actual - piso_origen))

        distancia = ascensor_usado.mover_a_piso(piso_origen)
        tiempo = distancia * tiempo_por_piso + tiempo_puertas
        total_tiempo_tradicional += tiempo

        # El ascensor se mueve al destino
        distancia_destino = ascensor_usado.mover_a_piso(piso_destino)
        total_tiempo_tradicional += distancia_destino * tiempo_por_piso + tiempo_puertas

    return total_tiempo_tradicional

# Simulación para el escenario con un ascensor prioritario
def simular_escenario_prioritario():
    # Crear ascensores en diferentes posiciones
    ascensores = [Ascensor(piso_actual=random.randint(1, 30)) for _ in range(3)]
    ascensor_prioritario = AscensorPrioritario(piso_actual=random.randint(1, 30))

    # Generar solicitudes con usuarios prioritarios
    solicitudes = generar_solicitudes_prioritarias(10, 3, 30, prioridad_ratio=0.2)

    # Ejecutar la simulación
    tiempo_total = algoritmo_look_multiascensor_prioritario(ascensores, ascensor_prioritario, solicitudes)

    print(f"Tiempo total con ascensor prioritario: {tiempo_total} segundos")

# Ejecutar la simulación con un ascensor prioritario
simular_escenario_prioritario()
```

Dos elevadores al mismo tiempo:

```python
class AscensorCoordinado(Ascensor):
    """
    Clase para ascensores coordinados que operan en el mismo eje vertical.
    Estos ascensores no pueden ocupar el mismo piso al mismo tiempo.
    """
    def __init__(self, piso_actual):
        super().__init__(piso_actual)

def verificar_colision(ascensor1, ascensor2, piso_destino):
    """
    Verifica si hay riesgo de colisión entre dos ascensores en el mismo eje.
    """
    if ascensor1.piso_actual == piso_destino and ascensor2.piso_actual == piso_destino:
        return True  # Colisión
    return False  # No hay colisión

def algoritmo_ascensores_mismo_eje(ascensor1, ascensor2, solicitudes):
    """
    Algoritmo para manejar dos ascensores en el mismo eje vertical.
    Se debe coordinar para evitar colisiones.
    """
    total_tiempo = 0

    # Procesamos las solicitudes
    for solicitud in solicitudes:
        piso_origen, piso_destino = solicitud

        # Verificar qué ascensor está más cercano al piso de origen
        if abs(ascensor1.piso_actual - piso_origen) < abs(ascensor2.piso_actual - piso_origen):
            ascensor_asignado = ascensor1
        else:
            ascensor_asignado = ascensor2

        # Verificar colisión antes de mover el ascensor
        if verificar_colision(ascensor1, ascensor2, piso_origen):
            # Si hay colisión, dar prioridad al que esté más cercano
            if ascensor_asignado == ascensor1:
                ascensor2.mover_a_piso(piso_origen + 1)  # Mover temporalmente el ascensor 2
            else:
                ascensor1.mover_a_piso(piso_origen - 1)  # Mover temporalmente el ascensor 1

        # Mover el ascensor asignado al piso de origen
        distancia_origen = ascensor_asignado.mover_a_piso(piso_origen)
        tiempo_origen = distancia_origen * tiempo_por_piso + tiempo_puertas
        total_tiempo += tiempo_origen

        # Mover al destino
        distancia_destino = ascensor_asignado.mover_a_piso(piso_destino)
        tiempo_destino = distancia_destino * tiempo_por_piso + tiempo_puertas
        total_tiempo += tiempo_destino

    return total_tiempo

# Simulación para el escenario con dos ascensores en el mismo eje vertical
def simular_ascensores_mismo_eje():
    # Crear dos ascensores en diferentes posiciones en el mismo eje
    ascensor1 = AscensorCoordinado(piso_actual=random.randint(1, 30))
    ascensor2 = AscensorCoordinado(piso_actual=random.randint(1, 30))

    # Generar solicitudes
    solicitudes = generar_solicitudes(10, 3, 20)

    # Ejecutar la simulación
    tiempo_total = algoritmo_ascensores_mismo_eje(ascensor1, ascensor2, solicitudes)

    print(f"Tiempo total con dos ascensores en el mismo eje: {tiempo_total} segundos")

# Ejecutar la simulación con dos ascensores en el mismo eje
simular_ascensores_mismo_eje()
```

Pruebas Adicionales 1 - Sobrecarga de solicitudes

```python
# Sobrecarga de solicitudes desde diferentes pisos al mismo tiempo
def prueba_sobrecarga_solicitudes():
    # Crear ascensores en diferentes posiciones
    ascensores = [AscensorConCapacidad(piso_actual=random.randint(1, 30)) for _ in range(3)]

    # Generar una gran cantidad de solicitudes simultáneas desde diferentes pisos
    solicitudes = generar_solicitudes_con_personas(15, 3, 50)  # Simulamos 50 personas en un mismo momento

    # Ejecutar la simulación
    tiempo_total = algoritmo_ascensor_con_capacidad(ascensores, solicitudes)

    print(f"Tiempo total bajo sobrecarga de solicitudes: {tiempo_total} segundos")

# Ejecutar la prueba de sobrecarga
prueba_sobrecarga_solicitudes()
```

```python
def prueba_sobrecarga_incremental():
    # Crear ascensores en diferentes posiciones
    ascensores = [AscensorConCapacidad(piso_actual=random.randint(1, 30)) for _ in range(3)]

    max_solicitudes = 0
    umbral_sobrecarga = 5000  # Definimos un umbral de tiempo en segundos para considerar sobrecarga
    solicitudes_iniciales = 10
    incremento = 5

    while True:
        # Generar solicitudes incrementales
        solicitudes = generar_solicitudes_con_personas(15, 3, solicitudes_iniciales)

        # Ejecutar la simulación
        tiempo_total = algoritmo_ascensor_con_capacidad(ascensores, solicitudes)

        print(f"Número de solicitudes: {solicitudes_iniciales}, Tiempo total: {tiempo_total} segundos")

        # Verificar si el sistema se sobrecarga
        if tiempo_total > umbral_sobrecarga:
            print(f"Sobrecarga detectada con {solicitudes_iniciales} solicitudes.")
            max_solicitudes = solicitudes_iniciales - incremento
            break

        # Incrementar el número de solicitudes para la siguiente simulación
        solicitudes_iniciales += incremento

    return max_solicitudes

# Ejecutar la prueba de sobrecarga incremental
max_solicitudes = prueba_sobrecarga_incremental()
print(f"El número máximo de solicitudes antes de sobrecargar el sistema es: {max_solicitudes}")
```

Interfaz de control por piso

```python
class InterfazControl:
    def __init__(self, ascensores):
        """
        Inicializa la interfaz de control con los ascensores disponibles.
        """
        self.ascensores = ascensores

    def seleccionar_destino(self, piso_actual, piso_destino):
        """
        Simula la selección de destino por parte del usuario.
        Calcula el ascensor más cercano y proporciona una estimación de tiempo.
        """
        # Seleccionar el ascensor más cercano al piso actual
        ascensor_asignado = min(self.ascensores, key=lambda asc: abs(asc.piso_actual - piso_actual))

        # Calcular el tiempo que tardará el ascensor en llegar al piso actual del usuario
        tiempo_llegada = abs(ascensor_asignado.piso_actual - piso_actual) * tiempo_por_piso + tiempo_puertas
        print(f"El ascensor llegará al piso {piso_actual} en aproximadamente {tiempo_llegada} segundos.")

        # Calcular el tiempo que tomará llegar al destino
        tiempo_viaje = abs(piso_destino - piso_actual) * tiempo_por_piso + tiempo_puertas
        tiempo_total = tiempo_llegada + tiempo_viaje
        print(f"El tiempo estimado total para llegar al piso {piso_destino} es de {tiempo_total} segundos.")

        # Simular el ascensor moviéndose
        ascensor_asignado.mover_a_piso(piso_actual)  # El ascensor llega al piso actual del usuario
        ascensor_asignado.mover_a_piso(piso_destino)  # Luego se mueve al destino

        return tiempo_total

# Simulación de la interfaz de control por piso
def simular_interfaz_control():
    # Crear ascensores en diferentes posiciones
    ascensores = [AscensorConCapacidad(piso_actual=random.randint(1, 30)) for _ in range(3)]

    # Inicializar la interfaz de control con los ascensores
    interfaz = InterfazControl(ascensores)

    # Ejemplo: El usuario está en el piso 10 y selecciona el piso 25 como destino
    piso_actual = 10
    piso_destino = 25

    # Usar la interfaz de control para seleccionar el destino y obtener la estimación de tiempo
    interfaz.seleccionar_destino(piso_actual, piso_destino)

# Ejecutar la simulación de la interfaz de control
simular_interfaz_control()
```
